```python
import os
import pandas as pd

directory_path = r"C:\Users\abedq\Desktop\Apziva\Apziva\P6\New folder\TIMIT Dataset\data\TRAIN\DR4\MMDM0"

# Create an empty dataframe with two columns: "name" and "content"
df = pd.DataFrame(columns=["name", "content"])

# Loop through all the files in the directory
for file_name in os.listdir(directory_path):
    # Check if the file is a .txt file
    if file_name.lower().endswith(".txt"):
        # Read the contents of the file
        file_path = os.path.join(directory_path, file_name)
        with open(file_path, "r") as f:
            content = f.read()
        
        # Add the file name and contents to the dataframe
        row = {"name": file_name, "content": content}
        df = df.append(row, ignore_index=True)

# Print the resulting dataframe
df
```
